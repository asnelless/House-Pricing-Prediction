```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
data.info()    #we can see that total_bedrooms value has null values
data.dropna(inplace=True)  #drops nan values and saves this option. now we have the same amount in all non-null values in all the diff fatures
```

```python
data.info()
```

```python
# split into training and testing(evaluate) data and split it into X and Y data 
from sklearn.model_selection import train_test_split

X = data.drop (['median_house_value'], axis =1)
Y = data['median_house_value']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2)
```

```python
train_data = X_train.join (Y_train)
#print(train_data) 
#print (train_data.hist())
```

```python
train_data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
train_data.hist (figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
# data preprocessing
# by taking logarithms let's see what will be the distribution
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_rooms']+1)
train_data['population'] = np.log(train_data['total_rooms']+1)
train_data['households'] = np.log(train_data['total_rooms']+1)

train_data.hist (figsize=(15,8)) 
```

```python
train_data.ocean_proximity.value_counts()   
```

```python
pd.get_dummies(train_data.ocean_proximity)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis = 1)
```

```python
# now let's see how the values correlate with the target variable
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
# coordinnates vizualization
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=train_data, hue = "median_house_value", palette = "coolwarm")
# the more red it gets, the more expensive the house is
# the more blue it gets, the cheaper the house is (the bottom line is coast/sea; the upper line is inland)
# the closer to the sea, the more expensive the house
```

```python
# feature engineering
# let's find out the number of bedrooms per rooms (how many of those rooms are bedrooms)
train_data['bedroom_relation'] = train_data ['total_bedrooms']/train_data['total_rooms']
# how many household rooms per rooms
train_data['household_rooms'] = train_data ['total_rooms']/train_data['households']

# the bedroom relation has a negative correlation with the median house value per block
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot = True, cmap = "YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# split into x and y
X_train, Y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s, Y_train)
```

```python
test_data = X_test.join (Y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_rooms']+1)
test_data['population'] = np.log(test_data['total_rooms']+1)
test_data['households'] = np.log(test_data['total_rooms']+1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis = 1)

test_data['bedroom_relation'] = test_data ['total_bedrooms']/test_data['total_rooms']
# how many household rooms per rooms
test_data['household_rooms'] = test_data ['total_rooms']/test_data['households']
```

```python
X_test, Y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform (X_test)
```

```python
reg.score(X_test_s, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor ()
forest.fit(X_train, Y_train)
```

```python
forest.score (X_test_s, Y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [100, 200, 300],
    "min_samples_split": [2,4],
    "max_depth": [None, 4, 8]
}

grid_search = GridSearchCV (forest, param_grid, cv=5,
                            scoring = "neg_mean_squared_error",
                            return_train_score=True)

grid_search.fit (X_train_s, Y_train)
```

```python
grid_search.best_estimator_
```
